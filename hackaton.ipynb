## ASM 계산식 사용하여 데이터 정리

ASM(appendicular skeletal muscle mass, 견갑골 근육량

ASM은 총 왼팔,오른팔, 왼다리, 오른다리의 근육량의 합으로 정의된다.

BIA_Right_Arm, BIA_Left_Arm, BIA_Right_Leg, BIA_Left_Leg 이용

ASM = BIA_Right_Arm + BIA_Left_Arm + BIA_Right_Leg + BIA_Left_Leg

ASM_Ht = ASM / (BIA_Height/100)^2

ASM_Wt : ASM / BIA_Weight

### T-score, Z-score(Using DEXA??)

T- score : 절대적인 위험도, 젊은 연령층(20대)의 골밀도와 비교한 점수(50세 이후 남성, 폐경 후 여성에서 적용)

Z- score : 이차성 원인 감별에 사용, 같은 연령대의 골밀도와 비교한 점수(50세 이전 남성, 폐경 전 여성에서 적용)

### 요추의 골밀도 해석

1) L1 ~ L4까지의 평균치를 진단  
2) 정상적으로 L1에서 L4로 가면서 골밀도가 증가하는데 (보통 음수이기 때문에 갈수록 0에 가까워져야 한다는 뜻) 이러한 경향이 역전되거나 T-score가 주위 척추와 1SD 이상의 차이를 나타나면 제외  
3) 2번으로 제외되는 척추골을 빼고 나머지 골밀도의 평균치로 해석

## 주요 용어 정리

ESR : 항응고제가 포함된 정맥혈을 시험관에 넣은 후 수직으로 세워 놓은 후 적혈구가 혈장으로부터 분리되어 시험관 아래로 가라앉는 시간당 이동거리를 의미. 면역질환이나 각종 염증질환, 감염, 외상 등에 있어서 환자 상태를 추적관찰
- 정상수치 : 0-10(남성), 0-20(여성)

CRP : 급성기 반응 물질로 감염이나 자가면역질환, 외상, 종양, 수술 등에 반응하여 증가하거나 감소하는 물질로 가장 민감한 반응을 보인다. 감염, 자가면역질환의 진단을 위한 도구가 아니라 치료 후 환자상태나 치료효과를 판단하는 자료로 사용한다.
- 정상수치 : 0.5-1.0

25-OH vitamin D : Vitamin D 결손진단, Vitamin D는 칼슘, 인 대사와 향상성을 조절하는 지용성 비타민으로 성장 발육 및 골 형성에 필수적
- 증가 : 원발성 부갑상선 기능 항진증, 말단 비대증, Vit D 과다증
- 감소 : 구루병, 골연화증, 만성 신부전
- 부족수치 : <10
- 불필요수치 : 10-29

CCP : 시트룰린을 인지하여 면역 시스템에 의해 생성되는 자가항체 단백으로 혈액에서 시트룰린 항체의 존재를 검출하는 검사에 이용된다. 비정상적인 아미노산인 시트룰린은 아미노산인 아르기닌이 변경되었을 때 생성되며 아르기닌에서 시트룰린으로의 전환이 류마티스 관절염 환자의 관절에서 관찰되는 자가 면역성 염증 과정에 역할을 하는 것으로 추측되고 있다. Anti-CCP는 민감도 및 특이도가 RF 보다 동등하거나 우수하고 류마티스 인자(RF)보다 훨씬 더 조기에 류마티스 관절염 환자에게서 양성을 보이는 것으로 나타났다. Anti-CCP는 증상의 시작 후 3-6개월 만에 초기 류마티스 환자의 약 50~60%에서 되며 류마티스 관절염의 조기 발견과 진단으로 의사들은 연관된 합병증과 조직의 손상을 최소화할 수 있도록 적극적인 치료를 시작할 수 있게 된다.
- 증가 : 류마티스 관절염

PET : 폐기능검사  
FVC : 안정된 상태에서 천천히 최대한 깊이 숨을 들이마신 후 최대한 빠르고 세게 내어 뱉는 공기량을 의미  
FEV1 : 숨을 최대로 들이 쉰 후 1초간 최대한 뱉을 수 있는 공기량을 의미  
FEV1/FVC : 기도폐쇄 유무를 확인하는 유용한 지표로 1초율이라고도 한다. 일반적으로 정상인은 FVC의 70%이상을 첫 1초에 내쉴 수 있다. 즉, 정상인은 70%이상이고, 그 이하인 경우에는 기도에 문제가 있다고 한다.

```python
from math import pi
from matplotlib.path import Path
from matplotlib.spines import Spine
from matplotlib.transforms import Affine2D
```

```python
## Make radar_chart
def radar_chart(insert_df,drawing_index):
    '''
    drawing_index input set (j,k)
    '''
    i,j = drawing_index
    labels = insert_df.columns
    num_labels = len(labels)

    angles = [x/float(num_labels)*(2*pi) for x in range(num_labels)] 
    angles += angles[:1] 

    my_palette = plt.cm.get_cmap("Dark2", len(range(i,j)))

    fig = plt.figure(figsize=(20,15))
    fig.set_facecolor('white')
    ax = fig.add_subplot(121,polar=True)
    for i, row in insert_df.iloc[i:j,:].iterrows():
        color = my_palette(i)
        data = insert_df.iloc[i].tolist()
        data += data[:1]

        ax.set_theta_offset(pi / 2) 
        ax.set_theta_direction(-1) 

        plt.xticks(angles[:-1], labels, fontsize=13) 
        ax.tick_params(axis='x', which='major', pad=15) 
        ax.set_rlabel_position(0) 
        plt.yticks([-2,-1,0,1,2,3],['-2','-1','0','1','2','3'], fontsize=10) 
        plt.ylim(-2,3)

        ax.plot(angles, data, color=color, linewidth=2, linestyle='solid', label=i) 
        ax.fill(angles, data, color=color, alpha=0.4) 

    for g in ax.yaxis.get_gridlines():
        g.get_path()._interpolation_steps = len(labels)

    spine = Spine(axes=ax,
              spine_type='circle',
              path=Path.unit_regular_polygon(len(labels)))


    spine.set_transform(Affine2D().scale(.5).translate(.5, .5)+ax.transAxes)

    ax.spines = {'polar':spine} 

    plt.legend(loc=(0.9,0.9))
    return fig,ax
```

### T-test(두 표본 평균 검정)
Levene 검정은 두 데이터의 분산 동일성을 검정 p-value가 클수록 분산이 비슷하다. $H_0 : $ 두 표본의 분산이 동일하지 않다.  
T 검정은 두 데이터의 평균이 같이 않다는 것을 검정, p-value <0.05 기준. $H_0 : $ 두 표본의 평균은 같다.

결과적으로 Right_Leg_BMD에서 유의미한 차이를 보인다.(Knee Rt Ap-stand 기준)

```python
def make_dt(row):
    lst=[]
    for i in association_df.columns.drop('RA'):
        if row[i] == 1:
            lst.append(i)
        elif i == 'ESR':
            if row[i] > 20:
                lst.append(i)
        elif i == 'CRP':
            if row[i] < 0.5:
                lst.append('CRP_decrease')
            elif row[i] >1.0:
                lst.append('CRP_increase')
        elif i == 'vit_D':
            if row[i] < 10:
                lst.append('vit_D_decrease')
            elif row[i] >29:
                lst.append('vit_D_increase')  
        elif i == 'CCP':
            if row[i] != 'Neg':
                lst.append(i)
        elif i == 'FEV1/FVC':
            if row[i] <70:
                lst.append('Pulmonary function error')
        elif i == 'RA':
            pass
    return lst
```

EMR 데이터에서 25명이상 질병을 가지는 Set를 가지고 연관분석을 실시

### 지지도(Support)
A 병에 걸린 환자 / 전체 환자

$support(A) = P(A)$

### 신뢰도(Confidence)
A 병이 있을 때, B 병도 있을 확률

$confidence(A -> B) = P(A \cap B)/P(A)$

### 향상도(Lift)
A 병을 갖고 있을 때 B 병도 있는지 연관성을 파악하는 비율

$Lift(A -> B) = P(A \cap B)/(P(A)P(B))$

1보다 값이 클 때, 연관성을 가진다고 한다.

```python
itemset = apriori(asso_df,min_support=0.1,use_colnames=True)
itemset
```

```python
## 신뢰도 >= 0.8
from mlxtend.frequent_patterns import association_rules
confidence_result = association_rules(itemset, metric="confidence", min_threshold=0.8)
confidence_result
```

```python
## Lift >= 1
association_rules(itemset, metric="lift", min_threshold=1) 
```

현재 데이터는 종속변수가 작으며 설명변수가 많이 들어가 있는 데이터이며 (250,138)의 형태를 가진다.

데이터가 적을 경우 resampling,bagging,boosting 등 다양한 기법이 존재한다.  
제가 선택한 방법은 간단한 모델 여러개를 사용하여 투표형식으로 예측한다.  
Voting Model의 경우 각 사용하는 모델이 독립적으로 예측하며, 그 결과들을 종합에서 하나의 예측값으로 산출하는 방법이다.  
Voting Regression model 과 Voting Classification model을 사용하였다.  
Regression의 경우 LinearRegression, RandomForestRegressor, LinearSVR을 사용하였고,  
Classification의 경우 LogisticRegression, RandomForestClassification, KNeighborsClassifier를 사용하였다.  
Target으로는 연관분석에서 사용한 EMS_DATA 변수를 Target으로 잡았다.  
또한 한 예측에서 여러가지 값을 multi-prediction을 수행하기 위해 ClassifierChain과 RegressorChain을 사용하였다.

```python
### regression part 와 classification part로 구분
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestRegressor, VotingRegressor, RandomForestClassifier, VotingClassifier
from sklearn.svm import LinearSVR
from sklearn.multioutput import ClassifierChain, RegressorChain
from sklearn.neighbors import KNeighborsClassifier
```

```python
## Classification
clf1 = LogisticRegression()
clf2 = RandomForestClassifier()
clf3 = KNeighborsClassifier()

eclf1 = VotingClassifier(estimators= [('lr',clf1),('rf',clf2),('knn',clf3)], voting='soft')
chain_eclf = ClassifierChain(eclf1,random_state=1)

## Regression
reg1 = LinearRegression()
reg2 = RandomForestRegressor()
reg3 = LinearSVR()

ereg1 = VotingRegressor(estimators=[('lr',reg1),('rf',reg2),('svr',reg3)])
chain_ereg = RegressorChain(ereg1,random_state=2)
```

```python
## Make radar_chart
def radar_chart_simulation(insert_df,drawing_index):
    '''
    drawing_index input set (j,k)
    '''
    j,k = drawing_index
    labels = insert_df.columns
    num_labels = len(labels)

    angles = [x/float(num_labels)*(2*pi) for x in range(num_labels)] 
    angles += angles[:1] 

    my_palette = plt.cm.get_cmap("Dark2", len(range(j,k)))

    fig = plt.figure(figsize=(20,15))
    fig.set_facecolor('white')
    ax = fig.add_subplot(121,polar=True)
    for i, row in insert_df.iloc[j:k,:].iterrows():
        color = my_palette(i)
        data = insert_df.iloc[i].tolist()
        data += data[:1]

        ax.set_theta_offset(pi / 2) 
        ax.set_theta_direction(-1) 

        plt.xticks(angles[:-1], labels, fontsize=13) 
        ax.tick_params(axis='x', which='major', pad=15) 
        ax.set_rlabel_position(0) 
        plt.yticks([-2,-1,0,1,2,3],['-2','-1','0','1','2','3'], fontsize=10) 
        plt.ylim(-2,3)

        ax.plot(angles, data, color=color, linewidth=2, linestyle='solid', label=i) 
#         ax.fill(angles, data, color=color, alpha=0.4) 

    for g in ax.yaxis.get_gridlines():
        g.get_path()._interpolation_steps = len(labels)

    spine = Spine(axes=ax,
              spine_type='circle',
              path=Path.unit_regular_polygon(len(labels)))


    spine.set_transform(Affine2D().scale(.5).translate(.5, .5)+ax.transAxes)

    ax.spines = {'polar':spine} 

    plt.legend(loc=(0.9,0.9))
    return fig,ax
```
